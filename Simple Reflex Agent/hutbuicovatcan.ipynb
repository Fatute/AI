```python
import numpy as np
import random
```

```python
def create_matrix():
    
    size_m = int(input("Nhập vào chiều dài sàn nhà (m): "))
    size_n = int(input("Nhập vào chiều rộng sàn nhà (n): "))
    
    matrix = np.random.randint(0, 3, (size_m, size_n))
    
    row = random.randrange(size_m)
    col = random.randrange(size_n)
    
    while (matrix[row, col] == 2):
        row = random.randrange(size_m)
        col = random.randrange(size_n)
    
    return matrix, row, col
```

```python
def valid_moves(matrix, row, col):
        
    list_of_directions = []
    size_mat = matrix.shape
    
    directions = [
        ['UP', -1, 0],
        ['DOWN', 1, 0],
        ['LEFT', 0, -1],
        ['RIGHT', 0, 1]
    ]
    
    for dir, x, y in directions:
        x_new = row + x
        y_new = col + y
        if x_new >= 0 and x_new < size_mat[0] and y_new >= 0 and y_new < size_mat[1] and matrix[x_new, y_new] != 2:
            list_of_directions.append([dir, x_new, y_new])
    
    return list_of_directions
```

```python
def run_reflex_agent(ini_matrix, count, row, col):

    cur_matrix = ini_matrix
    
    print("TRẠNG THÁI BAN ĐẦU")
    print(ini_matrix)
    print(f"MÁY HÚT BỤI ĐANG Ở VỊ TRÍ [{row}, {col}]")
    
    while True:
        print(f"Máy chạy lần thứ: {count + 1}\n")
        
        moves = valid_moves(cur_matrix, row, col)
        
        count += 1
        
        if count == 100:
            print("BẾ TẮC, MÁY NGỪNG!!!")
            break
        
        if len(moves) == 0:
            print("MÁY NGỪNG HOẠT ĐỘNG VÌ KHÔNG THỂ DI CHUYỂN!!!")
            break
        
        if cur_matrix[row, col] == 1:
            print("VỊ TRÍ NHÀ BẨN TIẾN HÀNH LÀM SẠCH\n")
            cur_matrix[row, col] = 0
            
        if 1 not in cur_matrix:
            print(cur_matrix)
            print("NHÀ ĐÃ SẠCH, MÁY NGỪNG LÀM VIỆC!!!")
            break
        
        direction, row, col = random.choice(moves)
        
        print(f"MÁY HÚT BỤI DI CHUYỂN {direction}, ĐẾN VỊ TRÍ [{row}, {col}]")
        
        print(cur_matrix)
```

```python
matrix_mxn, row, col = create_matrix()
run_reflex_agent(matrix_mxn, 0, row, col)
```

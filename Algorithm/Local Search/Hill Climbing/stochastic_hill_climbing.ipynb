```python
import numpy as np
import random
```

```python
def create_mat():
    
    new_mat = np.array([
        ['1', '1', '1', '0'],
        ['0', '0', 'x', '1'],
        ['1', '1', '0', '0'],
        ['1', '1', '0', '1']
    ])
    
    row, col = find_vacuum(new_mat)
    
    return new_mat, row, col
```

```python
def find_vacuum(matrix):
    pos = np.argwhere(matrix == 'x')[0]
    
    return pos[0], pos[1]
```

```python
def count_dirty_position(matrix):
    count = 0
    for x in range(4):
        for y in range(4):
            if matrix[x][y] == '1':
                count += 1
    return count
```

```python
def moves(matrix):
    
    lst_move = []
    row, col = find_vacuum(matrix)
    
    directions = [
        ['LEFT', 0, -1],
        ['RIGHT', 0, 1],
        ['UP', -1, 0],
        ['DOWN', 1, 0]
    ]
    
    for dir, x, y in directions:
        row_new = row + x
        col_new = col + y
        
        if row_new < 4 and row_new >= 0 and col_new < 4 and col_new >= 0:
            lst_move.append([row_new, col_new])
            
    return lst_move
```

```python
def stochastic(matrix, row, col):

    cur_state = matrix.copy()
    while True:

        h = count_dirty_position(cur_state)

        if h == 0:
            print("Nha da sach!!!")
            return

        print("Trang thai cha:")
        print(cur_state)
        print(f"h = {h}\n")
        
        lst_move = moves(cur_state)

        best_states = []
        
        print("Cac trang thai con: ")
        for x, y in lst_move:

            next_state = cur_state.copy()

            next_state[row][col] = '0'
            next_state[x][y] = 'x'

            h_new = count_dirty_position(next_state)

            print(next_state)
            print(f"h = {h_new}\n")
            
            if h_new < h:
                best_states.append((next_state, x, y))
        
        if len(best_states) == 0:
            print("Khong ton tai cach giai!!!")
            return
        
        chosen = random.choice(best_states)
        cur_state = chosen[0]
        row = chosen[1]
        col = chosen[2]
```

```python
if __name__ == "__main__":
    mat, row, col = create_mat()
    stochastic(mat, row, col)
```

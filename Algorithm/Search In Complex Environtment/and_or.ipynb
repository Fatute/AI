```python
import numpy as np
import random
```

```python
def find_vaccuum_position(matrix):
    pos = np.argwhere(matrix == 'x')
    if len(pos) == 0:
        return 0, 0
    return pos[0][0], pos[0][1]
```

```python
def moves(matrix):
    
    lst_move = []
    row, col = find_vaccuum_position(matrix)
    
    directions = [
        ['LEFT', 0, -1],
        ['RIGHT', 0, 1],
        ['UP', -1, 0],
        ['DOWN', 1, 0]
    ]
    
    for dir, x, y in directions:
        row_new = row + x
        col_new = col + y
        
        if row_new < 4 and row_new >= 0 and col_new < 4 and col_new >= 0:
            lst_move.append([dir, row_new, col_new])
            
    return lst_move
```

```python
def result_action(matrix, row, col, action):
    temp_mat = matrix.copy()
    _, dr, dc = action
    
    row_new, col_new = row + dr, col + dc

    temp_mat[row, col] = '0'
    temp_mat[row_new, col_new] = 'x'
     
    return temp_mat
```

```python
def matrix_to_tuple(matrix):
    return tuple(map(tuple, matrix))
```

```python
def or_search(state, path):
    if '1' not in state:
        return []
    
    state_tuple = matrix_to_tuple(state)
    if state_tuple in path:
        return False
    
    actions = moves(state)
    row, col = find_vaccuum_position(state)
    
    for act in actions:
        result_state = result_action(state, row, col, act)
        possible_states = [result_state]
        
        lucky = random.random()
        if lucky > 0.7:
            possible_states.append(np.copy(state))
        
        new_path = path + [state_tuple]
        plan = and_search(possible_states, new_path)
        
        if plan != False:
            return [act[0], plan]
            
    return False
```

```python
def and_search(states, path):
    plans = {}
    for s in states:
        s_tuple = tuple(map(tuple, s))
        
        plan_x = or_search(s, path)
    
        if plan_x == False:
            return False
    
        plans[s_tuple] = plan_x
    
    return plans
```

```python
def and_or_search(matrix):
    return or_search(matrix, [])
```

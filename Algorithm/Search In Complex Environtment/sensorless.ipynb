```python
import numpy as np
import random
import sys

sys.path.append("../")
from Node import Node
```

```python
def create_start():
    S = []
    for i in range(0, 2):
        arr = np.random.randint(0, 2, (4, 4))
        arr = arr.astype(str)
        
        x = random.randint(0, 3)
        y = random.randint(0, 3)
        
        arr[x, y] = 'x'
        S.append(arr)
        
    return S  
```

```python
def find_vaccuum_position(matrix):
    pos = np.argwhere(matrix == 'x')[0]
    
    return pos[0], pos[1]
```

```python
def moves(matrix):
    
    lst_move = []
    row, col = find_vaccuum_position(matrix)
    
    directions = [
        ['LEFT', 0, -1],
        ['RIGHT', 0, 1],
        ['UP', -1, 0],
        ['DOWN', 1, 0]
    ]
    
    for dir, x, y in directions:
        row_new = row + x
        col_new = col + y
        
        if row_new < 4 and row_new >= 0 and col_new < 4 and col_new >= 0:
            lst_move.append([dir, row_new, col_new])
            
    return lst_move
```

```python
def solution(node):
    path = []
    
    while node.parent is not None:
        path.append((node.action, node.state))
        
        node = node.parent
    
    path.reverse()
    
    return path
```

```python
def count_dirty(matrix):
    count = 0
    for x in range(0, 4):
        for y in range(0, 4):
            if matrix[x][y] == '1':
                count += 1
    return count
```

```python
def check_goal(lst):
    count = 0
    for x in lst:
        if count_dirty(x) == 0:
            count += 1
            
    if count == len(lst):
        return True
    
    return False
```

```python
def search_with_no_observation(start):

    print("TRANG THAI BAN DAU")

    for state in start:
        print(state)
        print()

    root = Node(start, None, None, 0)

    if check_goal(root.state):

        print("TIM THAY LOI GIAI")
        path = solution(node)
        print(f"Huong di chuyen: {path[0]}")
            
        for x in path:
            print(x[1])
        return
    
    frontier = [root]
    reached = []

    while frontier:

        node = frontier.pop()
        belief = node.state

        key = tuple(tuple(map(tuple, s))for s in belief)
        if key in reached:
            continue

        reached.append(key)
        if check_goal(belief):

            print("TIM THAY LOI GIAI")
            path = solution(node)
            print(f"Huong di chuyen: {path[0]}")
                
            for x in path:
                print(x[1])
            print('\n')
            return
        
        actions = set()

        for state in belief:
            for dir, _, _ in moves(state):
                actions.add(dir)

        for action in actions:

            new_belief = []
            for state in belief:
                child = state.copy()
                row, col = find_vaccuum_position(child)

                found = False

                for dir, x, y in moves(child):

                    if dir == action:

                        child[row, col] = '0'
                        child[x, y] = 'x'

                        found = True
                        break

                if not found:
                    child = state.copy()

                new_belief.append(child)

            child_key = tuple(tuple(map(tuple, s))for s in new_belief)

            if child_key not in reached:

                child_node = Node(new_belief, node, action, node.cost + 1)

                frontier.append(child_node)
    print("KHONG TIM THAY LOI GIAI")

    return None
```

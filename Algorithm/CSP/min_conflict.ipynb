```python
variables = [
    "Củ Chi",
    "Hóc Môn",
    "Quận 12",
    "Bình Tân",
    "Bình Chánh",
    "Gò Vấp",
    "Tân Bình",
    "Tân Phú",
    "Bình Thạnh",
    "Thủ Đức",
    "Phú Nhuận",
    "Quận 10",
    "Quận 11",
    "Quận 3",
    "Quận 1",
    "Quận 5",  
    "Quận 6",
    "Quận 4",
    "Quận 8",
    "Quận 7",
    "Nhà Bè",
    "Cần Giờ"
]

constrains = {
    "Củ Chi": ["Quận 12"],
    "Hóc Môn": ["Quận 12", "Bình Tân", "Bình Chánh", "Củ Chi"],
    "Quận 12": ["Gò Vấp", "Hóc Môn", "Tân Bình", "Tân Phú", "Bình Tân", "Thủ Đức", "Bình Thạnh"],
    "Bình Tân": ["Hóc Môn", "Quận 12", "Tân Phú", "Quận 6", "Quận 8", "Bình Chánh"],
    "Bình Chánh": ["Hóc Môn", "Bình Tân", "Quận 7", "Quận 8", "Nhà Bè"],
    "Gò Vấp": ["Quận 12", "Phú Nhuận", "Tân Bình", "Bình Thạnh", "Thủ Đức"],
    "Tân Bình": ["Quận 12", "Phú Nhuận", "Quận 11", "Quận 10", "Bình Tân", "Quận 3", "Quận 6"],
    "Tân Phú": ["Quận 12", "Bình Tân", "Quận 11", "Quận 6"],
    "Bình Thạnh": ["Quận 12", "Gò Vấp", "Phú Nhuận", "Quận 1", "Quận 4", "Thủ Đức"],
    "Thủ Đức": ["Quận 12", "Gò Vấp", "Phú Nhuận", "Bình Thạnh", "Quận 1", "Quận 4", "Quận 7"],
    "Phú Nhuận": ["Gò Vấp", "Tân Bình", "Quận 3", "Quận 1", "Thủ Đức", "Bình Thạnh"],
    "Quận 10": ["Quận 11", "Quận 5", "Tân Bình", "Quận 3", "Quận 4"],
    "Quận 11": ["Quận 10", "Tân Bình", "Tân Phú", "Quận 6", "Quận 4"],
    "Quận 3": ["Tân Bình", "Phú Nhuận", "Quận 5", "Quận 1", "Quận 10"],
    "Quận 1": ["Thủ Đức", "Bình Thạnh", "Phú Nhuận", "Quận 3", "Quận 5", "Quận 4"],
    "Quận 5": ["Quận 1", "Quận 3", "Quận 6", "Quận 8", "Quận 10", "Quận 11"],
    "Quận 6": ["Bình Tân", "Tân Bình", "Tân Phú", "Quận 11", "Quận 5", "Quận 8"],
    "Quận 4": ["Bình Thạnh", "Thủ Đức", "Quận 1", "Quận 7", "Quận 10", "Quận 11"],
    "Quận 8": ["Bình Tân", "Bình Chánh", "Quận 5", "Quận 6", "Quận 7"],
    "Quận 7": ["Quận 8", "Quận 4", "Thủ Đức", "Bình Chánh", "Nhà Bè"],
    "Nhà Bè": ["Cần Giờ", "Quận 7", "Bình Chánh"],
    "Cần Giờ": ["Nhà Bè"]
}

for node, neighbors in constrains.items():
    for neighbor in neighbors:
        if node not in constrains[neighbor]:
            constrains[neighbor].append(node)

colors = ["Đỏ", "Vàng", "Xanh lá", "Xanh Dương"]
keys = list(constrains.keys())
domains = {area: list(colors) for area in keys}
```

```python
def count_conflicts(var, value, current, constraints):
    conflicts = 0
    for neighbor in constraints[var]:
        if current[neighbor] == value:
            conflicts += 1
    return conflicts
```

```python
def check_all_constraints(assignment, constraints):
    is_valid = True
    for area, color in assignment.items():
        for neighbor in constraints[area]:
            if assignment[neighbor] == color:
                print(f"LỖI: '{area}' và '{neighbor}' kề nhau cùng màu '{color}'!")
                is_valid = False
    if is_valid:
        print("To mau thanh cong.")
```

```python
def get_conflicted_variables(current, constraints):
    conflicted_vars = []
    for var in current:
        for neighbor in constraints[var]:
            if current[var] == current[neighbor]:
                conflicted_vars.append(var)
                break 
    return conflicted_vars
```

```python
def minimize_conflicts(var, current, constraints, domains):

    min_conflicts = float('inf')
    best_values = []
    
    for value in domains[var]:
        conflicts = count_conflicts(var, value, current, constraints)
        if conflicts < min_conflicts:
            min_conflicts = conflicts
            best_values = [value]
        elif conflicts == min_conflicts:
            best_values.append(value)
            
    return random.choice(best_values)
```

```python
def min_conflict(variables, domains, constraints, max_steps):

    current = {var: random.choice(domains[var]) for var in variables}
    print(f"Trạng thái khởi tạo ban đầu: {current}")
    
    for i in range(1, max_steps + 1):
        
        conflicted_vars = get_conflicted_variables(current, constraints)
        
        if not conflicted_vars: 
            print(f"-> Đã tìm thấy nghiệm sau {i-1} bước lặp!")
            return current
        
        var = random.choice(conflicted_vars)

        value = minimize_conflicts(var, current, constraints, domains)
    
        current[var] = value

    return "failure"
```

```python
max_steps = 100

print("BẮT ĐẦU CHẠY THUẬT TOÁN MIN-CONFLICTS...")

result = min_conflict(variables, domains, constrains, max_steps)

print("\n--- KẾT QUẢ CUỐI CÙNG ---")
if result == "failure":
    print("Thất bại! Vượt quá số bước cho phép.")
else:
    for area, color in result.items():
        print(f"{area}: {color}")
    
    check_all_constraints(result, constrains)
```

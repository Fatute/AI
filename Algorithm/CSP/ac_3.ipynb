```python
def rm_inconsistance_value(X, Y, domains):
    removed = False
    for k in domains[X]:
        state = False
        for v in domains[Y]:
            if k != v:
                state = True
                break
        
        if not state:
            print(f"Voi X: {k}. Khong ton tai Y thoa man!!!")
            domains[X].remove(k)
            removed = True
        else:
            print(f"Voi X: {k}. Ton tai 1 Y thoa man!!!")
    
    return removed    
```

```python
def AC_3(constrains, queue, domains):
    
    while len(queue) != 0:
        X, Y = queue.pop(0)
        print(f"Xet cung: ({X}, {Y})")
        
        if rm_inconsistance_value(X, Y, domains):
            if len(domains[X]) == 0:
                return False
            for neighbor in constrains[X]:
                if neighbor != Y:
                    queue.append((neighbor, X))
    return True
```

```python
import copy

def backtrack_with_ac_3(assignment, variables, constrains, domains):
    if len(assignment) == len(variables):
        return assignment
    
    unassigned = [v for v in variables if v not in assignment]
    var = unassigned[0]
    
    for color in domains[var]:
        local_domains = copy.deepcopy(domains)
        
        local_domains[var] = [color]
        
        queue = []
        for neighbor in constrains[var]:
            if neighbor not in assignment:
                queue.append((neighbor, var))

        if AC_3(constrains, queue, local_domains):
            assignment[var] = color
            
            result = backtrack_with_ac_3(assignment, variables, constrains, local_domains)
            if result is not None:
                return result

            del assignment[var]
            
    return None
```

```python
def find_blank(matrix):
    blank = []
    for x in range(0, 3):
        for y in range(0, 3):
            if matrix[x][y] == "":
                blank.append((x, y))
    return blank
```

```python
def check_winner(matrix):
    for row in matrix:
        if row[0] == row[1] == row[2] != "":
            return row[0]
        
    for col in range(0, 3):
        if matrix[0][col] == matrix[1][col] == matrix[2][col] != "":
            return matrix[0][col]
        
    if matrix[0][0] == matrix[1][1] == matrix[2][2] != "":
        return matrix[0][0]
    
    if matrix[0][2] == matrix[1][1] == matrix[2][0] != "":
        return matrix[0][2]
    
    return False
```

```python
def utility(matrix):
    
    winner = check_winner(matrix)
    
    if winner == 'X':
        return 1
    
    if winner == 'O':
        return -1
    
    return 0
```

```python
def current_player(board):
    x_count = 0
    o_count = 0

    for row in board:
        for cell in row:
            if cell == 'X':
                x_count += 1
            elif cell == 'O':
                o_count += 1

    if x_count == o_count:
        return "X"
    else:
        return "O"
```

```python
def result(board, row, col, sign):
    temp = [r[:] for r in board]
    temp[row][col] = sign
    return temp
```

```python
def expectimax(board):

    global nodes
    nodes += 1

    if len(find_blank(board)) == 0 or check_winner(board) is not False:
        return utility(board), []

    turn = current_player(board)

    if turn == 'X':

        value = -float('inf')
        best_path = []

        for move in find_blank(board):

            child = result(board, move[0], move[1], 'X')
            score, path = expectimax(child)

            if score > value:
                value = score
                best_path = [move] + path

        return value, best_path

    else:

        blank_position = find_blank(board)

        expected_value = 0
        best_path = []

        rate = 1 / len(blank_position)

        for move in blank_position:

            child = result(board, move[0], move[1], 'O')
            score, path = expectimax(child)
            expected_value += rate * score

            if best_path == []:
                best_path = [move] + path

        return expected_value, best_path
```

```python
board = [
    ['', 'O', 'X'],
    ['', 'O', ''],
    ['X', '', '']
]
nodes = 0
print("Diem so, duong di tot nhat tim duoc", expectimax(board))
print("Tong so node duyet: ", nodes)
```
